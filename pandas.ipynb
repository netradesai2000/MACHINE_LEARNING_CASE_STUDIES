```python
'''
75% and above A geade
74-60----B
59-35---C
below 35---fail
p=marks/total*100
'''
#Function to calculate the grade---
def grade(marks):
  p=(marks/200*100)
  if(p>=75):
    return 'A'
  elif(p>60 and p<75):
    return 'B'
  elif(p>=35 and p<60):
    return 'C'
  else:
    return 'F'
```

```python
df['Grade']=df['Total'].apply(grade)
df
```
